```python
# –§—É–Ω–∫—Ü–∏—è –∏–º–ø–æ—Ä—Ç–∞ –∏–∑ *.csv
def load_from_csv(source1, source2, path2='https://code.s3.yandex.net/datasets/',
                  index_col=None, parse_dates=None, sep=','):
    """
        –§—É–Ω–∫—Ü–∏—è –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏–∏ –∑–∞–≥—Ä—É–∑–∫–∏ –¥–∞–Ω–Ω—ã—Ö –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∏–∑ —Ñ–∞–π–ª–∞ *.csv
        –Ω–∞ –≤—Ö–æ–¥–µ:
            source1 - –ø—É—Ç—å + —Ñ–∞–π–ª *.csv –Ω–∞ –ª–æ–∫–∞–ª—å–Ω–æ–º –∫–æ–º–ø—å—é—Ç–µ—Ä–µ
            source2 - —Ñ–∞–π–ª *.csv
            path2 - –ø—É—Ç—å (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 'https://code.s3.yandex.net/datasets/')
            sep - —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é ',')
        –Ω–∞ –≤—ã—Ö–æ–¥–µ:
            df (DataFrame)
    """
    try:
        df = pd.read_csv(source1, index_col=index_col, parse_dates=parse_dates, sep=sep)
    except ValueError:
        print('Fail - ValueError!')
    except Exception:
        print('Fail - –∑–∞–≥—Ä—É–∑–∫–∞ –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞')
        print('–ü–æ–ø—ã—Ç–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ –∏–∑ —Ä–µ–∑–µ—Ä–≤–Ω–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞')
        df = pd.read_csv(str(path2+source2), index_col=index_col, parse_dates=parse_dates, sep=sep)
        print(f'OK - –∑–∞–≥—Ä—É–∑–∫–∞ –∏–∑ —Ä–µ–∑–µ—Ä–≤–Ω–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞: "{str(path2+source2)}"')
    else:
        print(f'OK - –∑–∞–≥—Ä—É–∑–∫–∞ –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞: "{source1}"')
    finally:
        print('–ö–æ–Ω–µ—Ü.') 
    return df

#df = df.drop([''], axis=1)
#df.drop(index=index_list, inplace=True)
#df.to_csv('*.csv', index=False)
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Å—Ç–æ–ª–±—Ü–µ
def column_value_test(ds, test='un', prnt=True):
    """
        –§—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Å—Ç–æ–ª–±—Ü–µ
        'un' - unique(); 'na' - isna(); 'du' - duplicated()
    """
    if prnt:
        if test == 'un':
            print(f'–ö–æ–ª-–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ: "{ds.name}": {len(ds.unique())}/{len(ds)}')
        elif test == 'na':
            print(f'–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ: "{ds.name}": {ds.isna().sum()}/{len(ds)}')
        elif test == 'du':
            print(f'–ö–æ–ª-–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ: "{ds.name}": {ds.duplicated().sum()}/{len(ds)}')
        else:
            print(f'Error!')
    else:
        if test == 'un':
            return len(ds.unique())
        elif test == 'na':
            return ds.isna().sum()
        elif test == 'du':
            return ds.duplicated().sum()     
    
#column_value_test(data.iloc[:, 0])
#column_value_test(data.iloc[:, 0], test='na')
#column_value_test(data.iloc[:, 0], test='du')
```

```python
# –§—É–Ω–∫—Ü–∏—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Ç–∏–ø–∞
def change_type(ds, new_type='int8'):   
    """
        –§—É–Ω—Ü–∏—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Ç–∏–ø–∞
        datetime - '%Y-%m-%dT%H:%M:%S'
        –¥—Ä—É–≥–∏–µ - 
    """      
    old_type = type(ds.values[0])
    if new_type == 'datetime':
        try:
            #ds = pd.to_datetime(ds, format='%Y-%m-%dT%H:%M:%S')
            ds = pd.to_datetime(ds, format='%Y-%m-%d')
            print(f'OK - {ds.name:<20}: ({len(ds)}) {type(ds.values[0])}')
        except:
            print(f'FAIL - {ds.name:<20}: {type(ds.values[0])}')
    elif new_type == 'datetime64':
        try:
            ds = pd.to_datetime(ds, new_type)
            print(f'OK - {ds.name:<20}: ({len(ds)}) {old_type} --> {type(ds.values[0])}')
        except:
            print(f'FAIL - {ds.name:<20}: {type(ds.values[0])}')            
    else:    
        try:
            ds = ds.astype(new_type)
            print(f'OK - {ds.name:<20}: ({len(ds)}) {old_type} --> {type(ds.values[0])}')
        except:
            print(f'FAIL - {ds.name:<20}: {type(ds.values[0])}')     
    
    return ds
```

```python
# –§—É–Ω–∫—Ü–∏—è –ø–µ—á–∞—Ç–∏ –º–∞—Ç—Ä–∏—Ü—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
def plot_correlation_matrix(df, graphWidth):
    #df = df.dropna('columns')
    df = df.dropna()
    # keep columns where there are more than 1 unique values
    df = df[[col for col in df if df[col].nunique() > 1]]
    if df.shape[1] < 2:
        print(f'The number of non-NaN or constant columns({df.shape[1]}) is less than 2')
        return
    corr = df.corr(method='spearman')
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    plt.suptitle(f'–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π (—É–ø—Ä–æ—â—ë–Ω–Ω–æ)', fontsize=18)
    plt.show()
```

```python
# –¥–∏–∞–≥—Ä–∞–º–º–∞ –∏ —Ä–∞—Å—á—ë—Ç —Ä–∞–∑–º–∞—Ö–∞
def box_calc_span(df, graph=True, table=True, title='–î–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞–∑–º–∞—Ö–∞'):  
    # –æ—Ç–±—Ä–æ—Å–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏
    df = df.dropna()
    df = df.select_dtypes(include=[np.number])
    l1 = []; l2 = []; l3 = []; l4 = []; l5 = []

    if graph:
        # —Å–µ—Ç–∫–∞ –¥–∏–∞–≥—Ä–∞–º–º —Ä–∞–∑–º–∞—Ö–∞
        names = df.columns
        cnt = len(names)
        fig, axes = plt.subplots(nrows=cnt, ncols=1)
        fig.set_figwidth(15)
        fig.set_figheight(4*cnt)
        
        if cnt > 1:
            for name, ax in zip(names, axes.flatten()):
                #pd.plotting.boxplot(data=df, column=name, fontsize=12, rot=0, ax=ax)
                sns.boxplot(data=df, x=name, orient='h', ax=ax)
        elif cnt == 1:
            sns.boxplot(data=df, x=names[0], orient='h')
       
        #print('== –î–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞–∑–º–∞—Ö–∞ ==')
        plt.suptitle(f'–î–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞–∑–º–∞—Ö–∞', fontsize=16)
        plt.show()
    
    for col in df.columns:         
        # —Ä–∞—Å—á—ë—Ç –∫–≤–∞–Ω—Ç–∏–ª–µ–π 
        q1 = round(np.quantile(df[col], 0.25), 1)
        q3 = round(np.quantile(df[col], 0.75), 1)
        delta_q3_q1 = round(q3-q1, 2)
        left_whisk = round(q1 - 1.5*delta_q3_q1, 1)
        if df[col].min() > left_whisk:
            left_whisk = df[col].min()
        right_whisk = round(q3 + 1.5*delta_q3_q1, 1)
        if df[col].max() < right_whisk:
            right_whisk = df[col].max()        
        l1.append(q1); l2.append(q3); l3.append(delta_q3_q1); l4.append(left_whisk); l5.append(right_whisk) 
        
    lst = [l1, l2, l3, l4, l5]
    df_subs = pd.DataFrame(lst, index=['q1', 'q3', 'delta_q3_q1', 'left_whisk', 'right_whisk'],
                           columns=df.columns)
    if table:
        output = ""
        output = df_subs.style.format('{:.1f}').set_table_attributes("style='display:inline'")\
                .set_caption(f'== –î–∞–Ω–Ω—ã–µ —Ä–∞–∑–º–∞—Ö–∞ ==')._repr_html_()
        display(HTML(output))
    
    return df_subs
```

```python
import math

# –§—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ –≥—Ä—É–ø–ø—ã –∫—Ä—É–≥–æ–≤—ã—Ö –¥–∏–∞–≥—Ä–∞–º–º
def plot_pie(df):
    # –æ—Ç–±—Ä–æ—Å–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏
    df = df.dropna()
    # –≤—ã–±–æ—Ä —Å—Ç–æ–ª–±—Ü–æ–≤ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    cols = df[[col for col in df if ((df[col].nunique() > 1) & (df[col].nunique() < 12))]].columns
    # —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ count(cols)/3 x 3
    fig, axes = plt.subplots(math.ceil(len(cols)/3), 3, figsize=(12, 9))
    axes = axes.flat
    
    for i, col in enumerate(cols):
        values = df[col].value_counts()
        vpct = df[col].value_counts(normalize=True)
        first_small = len(values)
        sum = 0
        lst = []
        if ((vpct.min() < 0.03) & (vpct[:-1].min() < 0.03)):
            for k in range(len(vpct)):
                if vpct[k] < 0.03:
                    if first_small == len(values):
                        first_small = k
                    sum += values[k]
                    continue
                lst.append(k)

            values[first_small] = sum
            values = values[:first_small+1]
            lst = lst[:first_small]
            lst.append('–ü—Ä–æ—á–∏–µ')
            values = values.set_axis(lst)
        
        axes[i].pie(values, labels=values.index, autopct='%.0f%%', explode=[0.02]*values.size, radius=1)          
        axes[i].set_title(col)
    
    # —É–¥–∞–ª–µ–Ω–∏–µ –Ω–µ–∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω—ã—Ö –æ—Å–µ–π
    for i in range(1, 4 - (len(cols) % 3)):
        axes[-i].axis('off')
    
    plt.suptitle('–ö—Ä—É–≥–æ–≤—ã–µ –¥–∏–∞–≥—Ä–∞–º–º—ã', fontsize=16)
    plt.show()
```

```python
# –§—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ –≥—Ä—É–ø–ø—ã –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º
def plot_hist(df, item_to_width=5):
    # –æ—Ç–±—Ä–æ—Å–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏
    df = df.dropna()
    df = df.select_dtypes(include=[np.number])
    cols = df.columns
    k = 1
    if df.shape[1] < 5:
        df.hist(figsize=(10,10))
    else:
        item_to_height = math.ceil(len(cols) / item_to_width)
        k = item_to_height // item_to_width
        if k == 0:
            k = 1
        fig = df.hist(figsize=(100,100*k), bins=30, layout=(item_to_height, item_to_width),
                      xlabelsize=50, ylabelsize=50)
        [x.title.set_size(50) for x in fig.ravel()]
    #print('== –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã ==')
    if k == 1:
        plt.suptitle('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã', fontsize=20)
    else:
        plt.suptitle('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã', fontsize=k*25)
    plt.show()
```

```python
# –§—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ –≥—Ä—É–ø–ø—ã –¥–∏–∞–≥—Ä–∞–º–º —Ä–∞—Å—Å–µ—è–Ω–∏—è –∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π
def plot_scatter_matrix(df, plotSize, textSize):
    # –æ—Ç–±—Ä–æ—Å–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏
    df = df.dropna()
    # –æ—Å—Ç–∞–≤–∏—Ç—å —á–∏—Å–ª–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã
    df = df.select_dtypes(include=[np.number])
    # –æ—Å—Ç–∞–≤–∏—Ç—å —Å—Ç–æ–ª–±—Ü—ã —Å –∑–∞–¥–∞–Ω–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    df = df[[col for col in df if df[col].nunique() > 1]]
    columnNames = list(df)
    cN_len = len(columnNames)
    # –≤—ã–≤–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –ø–æ—Å—Ç—Ä–∞–Ω–∏—á–Ω–æ (10 –Ω–∞ —Å—Ç—Ä–∞–Ω–∏—Ü–µ)
    n = 11
    for cnt in range(0, cN_len, n):
        cN = columnNames[cnt:cnt+n]
        df_ = df[cN]
        ax = pd.plotting.scatter_matrix(df_, alpha=0.75, figsize=[plotSize, plotSize],
                                        diagonal='hist')
        # –ø—Ä–æ–ø–∏—Å—ã–≤–∞–Ω–∏–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ
        corrs = df.corr().values
        for i, j in zip(*plt.np.triu_indices_from(ax, k = 1)):
            ax[i, j].annotate('Corr. coef = %.3f' % corrs[i, j], (0.8, 0.2),
                              xycoords='axes fraction', ha='right', va='center', size=textSize)
        if cnt+10 <= len(columnNames):
            plt.suptitle(f'–î–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è –∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π {cnt+10}/{len(columnNames)}',
                         fontsize=32)
        else:
             plt.suptitle(f'–î–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è –∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–µ–π {len(columnNames)}/{len(columnNames)}',
                         fontsize=32)
        # –æ—Ç—Å—Ç—É–ø –ø–æ–¥–∑–∞–≥–æ–ª–æ–≤–∫–æ–≤ –æ—Ç –∑–∞–≥–æ–ª–æ–≤–∫–∞
        # plt.subplots_adjust(top=0)
    plt.show()
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º (un, na, du)
def col_info(df, un=True, na=True, du=True, title='–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º'):
    df = df.dropna()
    col_un = []; col_na = []; col_du = []
    for col in df.columns:
        if un:
            col_un.append(column_value_test(df[col], test='un', prnt=False))
        if na:
            col_na.append(column_value_test(df[col], test='na', prnt=False))
        if du:
            col_du.append(column_value_test(df[col], test='du', prnt=False))
        
    index = ['–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ', '–ü—Ä–æ–ø—É—Å–∫–∏', '–î—É–±–ª–∏–∫–∞—Ç—ã']
    result = pd.DataFrame(columns=df.columns, index=index, data=[col_un, col_na, col_du])
    output = ""
    output = result.style.set_table_attributes("style='display:inline'")\
            .set_caption(f'{title}')._repr_html_()
    display(HTML(output))
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –ø–æ –¥—É–±–ª–∏–∫–∞—Ç–∞–º c—Ç—Ä–æ–∫
def double_data(df, num_row=7, title='–î—É–±–ª–∏ —Å—Ç—Ä–æ–∫'):
    """
    —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –ø–æ –¥—É–±–ª–∏–∫–∞—Ç–∞–º —Å—Ç—Ä–æ–∫
    
    """ 
    df = df.dropna()
    num_dbl = df.duplicated().sum()
    if num_dbl < num_row:
        num_row = num_dbl
    result = df[df.duplicated()].head(num_row)
    output = ""
    output = result.style.set_table_attributes("style='display:inline'")\
            .set_caption(f'{title} ({num_row} –∏–∑ {num_dbl})')._repr_html_()
    display(HTML(output))
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∞ —Ç–∞–±–ª–∏—Ü—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
def simple_matrix_correlation(df, title='–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π'):
    df = df.dropna()
    result = df.corr(method='spearman').style.format('{:.2f}').background_gradient('coolwarm')
    output = ""
    output = result.set_table_attributes("style='display:inline'")\
            .set_caption(f'== {title} ==')._repr_html_()
    display(HTML(output))
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è "–ø–µ—Ä–≤—ã–π –≤–∑–≥–ª—è–¥ –Ω–∞ –¥–∞–Ω–Ω—ã–µ"
def first_view(df, vertical=False, corr_matrix=True, pie=True, hist=True, h_item_to_width=5, scatter=False): 
    """
    —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –æ–±—â–µ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
    
    """
    df.info();
    if vertical:
        display(form_vert_head(df.head(5)));
        display(df.iloc[:5,:2]);
        display(form_vert_head(df.describe()));
    else:
        display(df.head());
        display(df.describe().style.format('{:.1f}'));
    
    col_info(df)
    double_data(df)
    box_calc_span(df)
    
    if corr_matrix:
        plot_correlation_matrix(df, 20)
    
    if vertical:
        print('== –ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π ==')
        display(form_vert_head(df.corr(method='spearman')).background_gradient('coolwarm'));
    else:
        simple_matrix_correlation(df)
        
    if pie:
        plot_pie(df)  
    if hist:
        plot_hist(df, h_item_to_width)
    if scatter:
        plot_scatter_matrix(df, 30, 15)
```

```python
# –í—ã–≤–æ–¥ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–æ–≤ –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω–æ
from IPython.display import display, HTML

def display_side_by_side(dfs:list, captions:list):
    """Display tables side by side to save vertical space
    Input:
        dfs: list of pandas.DataFrame
        captions: list of table captions
    """
    output = ""
    combined = dict(zip(captions, dfs))
    for caption, df in combined.items():
        output += df.style.format('{:.2f}').set_table_attributes("style='display:inline'")\
            .set_caption(caption)._repr_html_()
        output += "\xa0\xa0\xa0"
    display(HTML(output))
```

```python
# –§—É–Ω–∫—Ü–∏—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (–æ–±–æ–±—â—ë–Ω–Ω–æ –¥–ª—è pe, sp, ke) –ø–æ –µ—ë —á–∏—Å–ª–æ–≤—ã–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º
def corr_analizer(row, name):
    """
        –§—É–Ω–∫—Ü–∏—è –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç —á–∏—Å–ª–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —à–∫–∞–ª–µ –ß–µ–¥–¥–æ–∫–∞ –∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –≤ —Å–ª–æ–≤–µ—Å–Ω—É—é —Ñ–æ—Ä–º—É
        –û—Ü–µ–Ω–∫–∞ —É–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω–∞ –¥–ª—è –º–µ—Ç–æ–¥–æ–≤ ("–ø–∏—Ä—Å–æ–Ω", "—Å–ø–∏—Ä–º–µ–Ω", "–∫–µ–Ω–¥–∞–ª–ª")
    """    
    n_sign = ['–Ω—É–ª–µ–≤–∞—è', '–æ—Ç—Ä–∏—Ü–∞—Ç.', '–ø–æ–ª–æ–∂–∏—Ç.']
    r_ch = ['–æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç', '–æ—á.—Å–ª–∞–±–∞—è', '—Å–ª–∞–±–∞—è', '—Å—Ä–µ–¥–Ω—è—è', '–≤—ã—Å–æ–∫–∞—è', '–æ—á.–≤—ã—Å–æ–∫–∞—è']
    r_ch_id = [0.1, 0.3, 0.5, 0.7, 0.9] # –ø–æ —à–∫–∞–ª–µ –ß–µ–¥–¥–æ–∫–∞
    
    value = row[name]
    if value == 0:
        row['-/+'] = '–Ω—É–ª–µ–≤–∞—è'
    elif value > 0:
        row['-/+'] = '–ø–æ–ª–æ–∂–∏—Ç.'
    elif value < 0:
        row['-/+'] = '–æ—Ç—Ä–∏—Ü–∞—Ç.'
    else:
        row['-/+'] = '–û—à–∏–±–∫–∞ –≤ –∑–Ω–∞–∫–µ!'
    
    if (abs(value) >= 0.0) & (abs(value) < 0.1):
        row['r'] = '–æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç'    
    elif (abs(value) >= 0.1) & (abs(value) < 0.3):
        row['r'] = '–æ—á.—Å–ª–∞–±–∞—è'
    elif (abs(value) >= 0.3) & (abs(value) < 0.5):
        row['r'] = '—Å–ª–∞–±–∞—è'
    elif (abs(value) >= 0.5) & (abs(value) < 0.7):
        row['r'] = '—Å—Ä–µ–¥–Ω—è—è'
    elif (abs(value) >= 0.7) & (abs(value) < 0.9):
        row['r'] = '—Å–∏–ª—å–Ω–∞—è'
    elif (abs(value) >= 0.9) & (abs(value) <= 1.0):
        row['r'] = '–æ—á.—Å–∏–ª—å–Ω–∞—è'
    else:
        row['-/+'] = '–û—à–∏–±–∫–∞ –≤ –≤–µ–ª–∏—á–∏–Ω–µ!'
        
    return row[['-/+', 'r']]

# –§—É–Ω–∫—Ü–∏—è —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç—á—ë—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
def corr_former(df, first, other, method_corr='pearson'):
    """
        –§—É–Ω–∫—Ü–∏—è —Ñ–æ—Ä–º–∏—Ä—É–µ—Ç —Ç–∞–±–ª–∏—á–Ω—ã–π –≤–∏–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ä–∞–±–æ—Ç—ã –º–µ—Ç–æ–¥–æ–≤ –∫–æ—Ä–µ–ª–ª—è—Ü–∏–∏
    """    
    corr_list = []
    for item in other:
        corr = round(df[first].corr(df[item], method=method_corr), 2)
        corr_list.append(corr)
    method_short = method_corr[:2]
    df_c = pd.DataFrame(columns=[method_short, first, '|', '-/+', '–°–≤—è–∑—å'])
    df_c[first] = corr_list
    df_c[method_short] = '|'
    df_c['|'] = '|'
    df_c[['-/+', '–°–≤—è–∑—å']] = df_c.apply(lambda x: corr_analizer(x, first), axis=1)
    df_c = df_c.rename(index=lambda x: other[x])

    display(df_c) 
```

```python
# –§—É–Ω–∫—Ü–∏—è —Ä–∞–∑–¥–µ–ª–µ–ª–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –≤—ã–±–æ—Ä–∫–∏: —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é, —Ç–µ—Å—Ç–æ–≤—É—é
def split_stratified_into_train_val_test(df_input, stratify_colname='y',
                                         frac_train=0.6, frac_valid=0.2, frac_test=0.2,
                                         random_state=None, shuffle=True, prnt=True):
    '''
    Splits a Pandas dataframe into three subsets (train, val, and test)
    following fractional ratios provided by the user, where each subset is
    stratified by the values in a specific column (that is, each subset has
    the same relative frequency of the values in the column). It performs this
    splitting by running train_test_split() twice.

    Parameters
    ----------
    df_input : Pandas dataframe
        Input dataframe to be split.
    stratify_colname : str
        The name of the column that will be used for stratification. Usually
        this column would be for the label.
    frac_train : float
    frac_valid : float
    frac_test  : float
        The ratios with which the dataframe will be split into train, val, and
        test data. The values should be expressed as float fractions and should
        sum to 1.0.
    random_state : int, None, or RandomStateInstance
        Value to be passed to train_test_split().

    Returns
    -------
    X_train, X_val, X_test, y_train, y_valid, y_test:
        Dataframes containing the six splits.
    '''

    if frac_train + frac_valid + frac_test != 1.0:
        raise ValueError('fractions %f, %f, %f do not add up to 1.0' % \
                         (frac_train, frac_valid, frac_test))

    if stratify_colname not in df_input.columns:
        raise ValueError('%s is not a column in the dataframe' % (stratify_colname))

    # Contains dataframe without stratify column
    X = df_input.drop([stratify_colname], axis=1)
    
    # Dataframe of just the column on which to stratify
    y = df_input[[stratify_colname]]

    if frac_valid > 0:
        # Split X and y into train and temp.
        X_train, X_temp, y_train, y_temp = train_test_split(X,
                                                            y,
                                                            #stratify=y,
                                                            test_size=(1.0 - frac_train),
                                                            random_state=random_state, shuffle=shuffle)

        # Split the temp X and y into val and test.
        relative_frac_test = frac_test / (frac_valid + frac_test)
        X_valid, X_test, y_valid, y_test = train_test_split(X_temp,
                                                            y_temp,
                                                            #stratify=y_temp,
                                                            test_size=relative_frac_test,
                                                            random_state=random_state, shuffle=shuffle)
    else:
        X_train, X_test, y_train, y_test = train_test_split(X,
                                                            y,
                                                            #stratify=y,
                                                            test_size=(1.0 - frac_train),
                                                            random_state=random_state, shuffle=shuffle)
        X_valid = {}
        y_valid = {}

    assert len(df_input) == len(X_train) + len(X_valid) + len(X_test)
    assert len(df_input) == len(y_train) + len(y_valid) + len(y_test)
    
    if prnt:
        print('dataframe_in:   ', df_input.shape)
        print('features_train: ', X_train.shape)
        print('features_test:  ', X_test.shape)
        if frac_valid > 0:
            print('features_valid:  ', X_valid.shape)
        print('target_train:   ', y_train.shape)
        print('target_test:    ', y_test.shape)
        if frac_valid > 0:
            print('target_valid:    ', y_valid.shape)
        
    if frac_valid > 0:
        return X_train.squeeze(), X_valid.squeeze(), X_test.squeeze(), \
               y_train.squeeze(), y_valid.squeeze(), y_test.squeeze()
    elif frac_valid == 0:
        return X_train.squeeze(), X_test.squeeze(), y_train.squeeze(), y_test.squeeze()
    else:
        return "Error!"
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –≤ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
def start_encoder(encoder, df, non_encoded_columns, flag='ohe'):   
    # –∏—Å—Ö–æ–¥–Ω–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ
    print('df_in: ', df.shape)
    
    # –∏–º–µ–Ω–∞ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    encoded_columns_names = encoder.get_feature_names_out(non_encoded_columns)
    
    # –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    encoded_columns = None
    if flag == 'ohe':    
        encoded_columns = encoder.transform(df[non_encoded_columns]).toarray()
    else:
        encoded_columns = encoder.transform(df[non_encoded_columns])    
    
    # —É–¥–∞–ª–µ–Ω–∏–µ –Ω–µ–∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    df.drop(non_encoded_columns, axis=1, inplace=True)
    
    # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
    for col, name in zip(encoded_columns.T, encoded_columns_names):
        df[name] = col
    
    # –∏—Ç–æ–≥–æ–≤–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ
    print('df_out:', df.shape)
    
    return df


# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –≤ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
# –û–±—ë—Ä—Ç–∫–∞ —Å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å—é —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–∏ –∫–æ–¥–∏—Ä–æ–≤—â–∏–∫–∞
def multi_encoder(encoder, df1_, df2_=None, df3_=None, flag1='ohe', flag2='3', flag3='fit'):
    # –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏–µ, –∏–Ω–∞—á–µ —Ñ—É–Ω–∫—Ü–∏—è –∏–∑–º–µ–Ω–∏—Ç –∏—Å—Ö–æ–¥–Ω—ã–π —Ñ–∞–π–ª
    df1 = df1_.copy()

    # –ø–æ–ª—É—á–∞–µ–º —Å–ø–∏—Å–æ–∫ –Ω–µ–∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    non_encoded_columns = [col for col in df1.columns if df1[col].dtype == 'object']
    
    if flag3 == 'fit':
        # –æ–±—É—á–µ–Ω–∏–µ encoder –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
        encoder.fit(df1[non_encoded_columns])
    
    df1 = start_encoder(encoder, df1, non_encoded_columns, flag=flag1)       

    if flag2 == '3':
        df2 = df2_.copy(); df3 = df3_.copy()
        df2 = start_encoder(encoder, df2, non_encoded_columns, flag=flag1)
        df3 = start_encoder(encoder, df3, non_encoded_columns, flag=flag1)
        return df1, df2, df3
    elif flag2 == '2':
        df2 = df2_.copy()
        df2 = start_encoder(encoder, df2, non_encoded_columns, flag=flag1)
        return df1, df2
    else:      
        return df1
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –∏ —Å—Ç—Ä–æ–∫
def drop(df, flag='columns', name=''):
    print('Dataframe_in: ', df.shape)
    if flag == 'columns':
        try:
            df.drop(name, axis=flag, inplace=True)
            print(f'–°—Ç–æ–ª–±–µ—Ü "{name}" —É–¥–∞–ª—ë–Ω.')
        except:
            print(f'–û—à–∏–±–∫–∞ —É–¥–∞–ª–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ "{name}".')
    elif flag == 'rows':
        try:
            df.drop(name, axis='rows', inplace=True)
            print(f'–°—Ç—Ä–æ–∫–∞ "{name}" —É–¥–∞–ª—ë–Ω–∞.')
        except:
            print(f'–û—à–∏–±–∫–∞ —É–¥–∞–ª–µ–Ω–∏—è —Å—Ç—Ä–æ–∫–∏ "{name}".')
    else:
        Print('Error!')
    print('Dataframe_out:', df.shape)
    print()
    return df
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è —É–¥–∞–ª–µ–Ω–∏—è —Å—Ç—Ä–æ–∫ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ
def dropna(df, name=''):
    print('dataframe_in: ', df.shape)
    try:
        df = df.dropna(subset=[name])
        print(f'–£–¥–∞–ª–µ–Ω—ã —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ "{name}".')
    except:
        print(f'–û—à–∏–±–∫–∞ —É–¥–∞–ª–µ–Ω–∏—è —Å—Ç—Ä–æ–∫ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ "{name}".')
    print('dataframe_out:', df.shape)
    print()
    return df
```

```python
# Initialize spacy 'en' model, keeping only tagger component needed for lemmatization
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])

# –æ–±—â–∏–π —Å–ª–æ–≤–∞—Ä—å —Å–ª–æ–≤
VOCABULARY = set()

# —Å–æ–∑–¥–∞–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è
def create_vocabulary(text):
    for word in text:
        VOCABULARY.add(word.lower())        

# –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—è
def lemmatize(text):
    lemma_list = [token.lemma_ for token in nlp(text)]
    #create_vocabulary(lemma_list)
    
    return ' '.join(lemma_list)
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –ø—Ä–æ–≥–Ω–æ–∑–∞ —Å —Ñ–∞–∫—Ç–æ–º
def graph_compare(target, predict, model_class, title=''):
    pd.DataFrame({ 
        'target': target,
        'predict': predict
    }).plot(figsize=(12,6), title=f'C—Ä–∞–≤–Ω–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–æ–≥–Ω–æ–∑–∞ —Å —Ñ–∞–∫—Ç–æ–º ({title}) {model_class}')
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫–∏ F1
def calc_F1(target, predict, prnt=False, title=''):
    f1 = f1_score(target, predict)
    if prnt:
        print(f'== {title} ==')
        print('F1:', f1)
        return
    return f1

f1_scorer = make_scorer(calc_F1, greater_is_better=True)
```

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html

```python
# —Ñ—É–Ω–∫—Ü–∏—è –∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –Ω–∞ RandomizedSearchCV
def learn_model_rscv(model, params, features, target, columns_cat, flag_details=True):  

    if model.__class__ == CatBoostClassifier:
        rscv = RandomizedSearchCV(model, param_distributions=params, n_jobs=-1, random_state=12345,
                                  cv=5, scoring='f1', verbose=False)
        rscv.fit(features, target, cat_features=columns_cat, verbose=False)
    elif model.__class__ == LGBMClassifier:
        rscv = RandomizedSearchCV(model, param_distributions=params, n_jobs=-1, random_state=12345,
                                  cv=5, scoring='f1', verbose=False)
        rscv.fit(features, target, categorical_feature=columns_cat)   
    else:
        rscv = RandomizedSearchCV(model, param_distributions=params, n_jobs=-1, random_state=12345,
                                  cv=5, scoring='f1', verbose=False)
        rscv.fit(features, target)
    
    means = rscv.cv_results_['mean_test_score']
    f1 = rscv.cv_results_['mean_test_score'][rscv.best_index_]
    params = rscv.cv_results_['params']
    #stds = rscv.cv_results_['std_test_score']

    time_fit = rscv.cv_results_['mean_fit_time'][rscv.best_index_]
    time_predict = rscv.cv_results_['mean_score_time'][rscv.best_index_]
    
    if flag_details:
        print('–ü–µ—Ä–µ–±–æ—Ä:')
        for mean, param in zip(means, params):
            print(f'-- RSCV RMSE: {round(mean, 3):<8} -- Params: {param}')
        print()
    
    print('–õ—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:')
    print(f'>> RSCV f1: {round(f1, 3):<8} >> Params: {rscv.best_params_}')
    print(f'>> RSCV Time fit: {round(time_fit, 2):<8}')
    print(f'>> RSCV Time predict: {round(time_predict, 2):<8}')
    print()
 
    return rscv
```

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

```python
# —Ñ—É–Ω–∫—Ü–∏—è –∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –Ω–∞ GridSearchCV
def learn_model_gscv(model, params, features, target, columns_cat, flag_details=True):  

    if model.__class__ == CatBoostClassifier:
        gscv = GridSearchCV(model, param_grid=params, n_jobs=-1,
                            cv=5, scoring='f1', verbose=False)
        gscv.fit(features, target, cat_features=columns_cat, verbose=False)
    elif model.__class__ == LGBMClassifier:
        gscv = GridSearchCV(model, param_grid=params, n_jobs=-1,
                            cv=5, scoring='f1', verbose=False)
        gscv.fit(features, target, categorical_feature=columns_cat)     
    else:
        gscv = GridSearchCV(model, param_grid=params, n_jobs=-1, 
                            cv=5, scoring='f1', verbose=False)
        gscv.fit(features, target)
    
    means = gscv.cv_results_['mean_test_score']
    f1 = gscv.cv_results_['mean_test_score'][gscv.best_index_]
    params = gscv.cv_results_['params']
    #stds = gscv.cv_results_['std_test_score']
    time_fit = gscv.cv_results_['mean_fit_time'][gscv.best_index_]
    time_predict = gscv.cv_results_['mean_score_time'][gscv.best_index_]
    
    if flag_details:
        print('–ü–µ—Ä–µ–±–æ—Ä:')
        for mean, param in zip(means, params):
            print(f'-- GSCV f1: {round(mean, 3):<8} -- Params: {param}')
        print()
    
    print('–õ—É—á—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:')
    print(f'>> GSCV f1: {round(f1, 3):<8} >> Params: {gscv.best_params_}')
    print(f'>> GSCV Time fit: {round(time_fit, 2):<8}')
    print(f'>> GSCV Time predict: {round(time_predict, 2):<8}')
    print()
 
    return gscv
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –∏–∑—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π:
# –ü–æ–ª—É—á–µ–Ω–∏–µ RMSE (–∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏—è), –≤—Ä–µ–º—è –æ–±—É—á–µ–Ω–∏—è (–∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏—è), –≤—Ä–µ–º—è –ø—Ä–µ–¥–∏–∫—Ü–∏–∏ (–∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏—è)
def learn_model(model, params, features1, target1, features2, target2, columns_cat=None,
                flag_compare=False, flag_scv='gscv', flag_details=True): 
   
    # "–®–∞–ø–∫–∞"
    print(f'== {model} ==')
    print(f'{params}')
    print()
    scv = None
    # –≤—ã–±–æ—Ä –ø–æ–∏—Å–∫–æ–≤–æ–≥–æ –∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ç–æ—Ä–∞
    if flag_scv == 'gscv':
        scv = learn_model_gscv(model, params, features1, target1, columns_cat,
                               flag_details=flag_details) 
    elif flag_scv == 'rscv':
        scv = learn_model_rscv(model, params, features1, target1, columns_cat,
                               flag_details=flag_details)
    else:
        print('Error of flag "scv"!')
        return
    
    # –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∏–∑ scv
    f1_cv = scv.cv_results_['mean_test_score'][scv.best_index_]
    time_fit = scv.cv_results_['mean_fit_time'][scv.best_index_]
    time_predict = scv.cv_results_['mean_score_time'][scv.best_index_]       

    predict2 = scv.best_estimator_.predict(features2)
    f1 = f1_score(target2, predict2)
    
    #print(f'{scv.best_params_}')
    print(f'Test f1:', round(f1, 3))
    print()
       
    # –≥—Ä–∞—Ñ–∏–∫ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –ø—Ä–æ–≥–Ω–æ–∑–∞ —Å —Ñ–∞–∫—Ç–æ–º
    if flag_compare:     
        graph_compare(target2, predict2, model.__class__, title='test')
    
    return {'f1_cv': f1_cv, 'f1': f1, 'time_fit': time_fit, 'time_pred': time_predict, 'scv': scv}
```

```python
# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Ç–æ–∫–µ–Ω–∏–∑–∞—Ç–æ—Ä–∞
tokenizer = tokenizer_class.from_pretrained('bert-base-uncased')
```

```python
# —Ç–æ–∫–µ–Ω–∏–∑–∞—Ü–∏—è —Ç–≤–∏—Ç–æ–≤
tokenized = df1['lemma_text'][:16000] \
    .progress_apply(lambda x: tokenizer.encode(x[:512], add_special_tokens=True, truncation=True))
```

–°–æ–≤–µ—Ç—É—é —Ç–µ–±–µ –∏—Å–ø–ª—å–∑–æ–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä truncation=True

```python
# 
lenght = len(tokenized[0])
lenght
```

```python
# –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –¥–ª–∏–Ω–∞ —Ç–æ–∫–µ–Ω–∞
max_lenght = max([len(i) for i in tokenized.values])
max_lenght
```

```python
# –≤–µ–∫—Ç–æ—Ä —Å –æ—Ç—Å—Ç—É–ø–∞–º–∏
padded = np.array([i + [0]*(512-len(i)) for i in tokenized.values])
len(padded[0])
```

```python
# –º–∞—Å–∫–∞ –¥–ª—è –≤—ã–¥–µ–ª–µ–Ω–∏—è –≤–∞–∂–Ω—ã—Ö —Ç–æ–∫–µ–Ω–æ–≤
attention_mask = np.where(padded != 0, 1, 0)
padded.shape, attention_mask.shape
```

<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞  <a class="tocSkip"> </h2>

<b>–í—Å–µ –æ—Ç–ª–∏—á–Ω–æ!üëç:</b>
    
padding —É–∫–∞–∑–∞–Ω, attention_mask –Ω–∞ –º–µ—Å—Ç–µ)

```python
# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–¥–µ–ª–∏
model = model_class.from_pretrained(pretrained_weights)
model = model.to(device)
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–æ–≤ –≤ —ç–º–±–µ–¥–¥–∏–Ω–≥–∏

# –≠–º–±–µ–¥–¥–∏–Ω–≥–∏ –º–æ–¥–µ–ª—å BERT —Å–æ–∑–¥–∞—ë—Ç –±–∞—Ç—á–∞–º–∏. –ß—Ç–æ–±—ã —Ö–≤–∞—Ç–∏–ª–æ –æ–ø–µ—Ä–∞—Ç–∏–≤–Ω–æ–π –ø–∞–º—è—Ç–∏, –±–∞—Ç—á –Ω–µ–±–æ–ª—å—à–æ–π:
batch_size = 64
# —Å–¥–µ–ª–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤ —Ç–≤–∏—Ç–æ–≤
embeddings = []

for i in tqdm(range(padded.shape[0] // batch_size)):
    # –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞–Ω–Ω—ã–µ
    batch = padded[batch_size * i:batch_size * (i + 1)]
    # –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –º–∞—Å–∫—É  
    attention_mask_batch = attention_mask[batch_size * i:batch_size * (i + 1)]
    # 
    seq_length = batch.shape[1]    
    # –î–ª—è —É—Å–∫–æ—Ä–µ–Ω–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–µ–π no_grad()
    # (–∞–Ω–≥–ª. no gradient, ¬´–Ω–µ—Ç –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞¬ª) –≤ –±–∏–±–ª–∏–æ—Ç–µ–∫–µ torch —É–∫–∞–∂–µ–º,
    # —á—Ç–æ –≥—Ä–∞–¥–∏–µ–Ω—Ç—ã –Ω–µ –Ω—É–∂–Ω—ã: –º–æ–¥–µ–ª—å BERT –æ–±—É—á–∞—Ç—å –Ω–µ –±—É–¥–µ–º.
    with torch.no_grad():
        batch = torch.LongTensor(batch).to(device)
        attention_mask_batch = torch.LongTensor(attention_mask_batch).to(device)        
        batch_embeddings = model(batch, attention_mask=attention_mask_batch)
    embeddings.append(batch_embeddings[0][:,0,:].cpu().numpy())
```

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

```python
# –§—É–Ω–∫—Ü–∏—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è ROC-–∫—Ä–∏–≤–æ–π –∑–∞–¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
def grafix_roc(fpr, tpr, title=''):
    plt.figure()
    plt.plot(fpr, tpr)
    # ROC-–∫—Ä–∏–≤–∞—è —Å–ª—É—á–∞–π–Ω–æ–π –º–æ–¥–µ–ª–∏ (–≤—ã–≥–ª—è–¥–∏—Ç –∫–∞–∫ –ø—Ä—è–º–∞—è)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC-–∫—Ä–∏–≤–∞—è ({title})')
    plt.grid()
    plt.show()
```

```python
def test_model1(model, features1, target1, features2, target2, title=''):    
    # –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å
    model.fit(features1, target1)
    
    # –ø–æ–¥–≥–æ—Ç–æ–≤–∏–º –ø—Ä–æ–≥–Ω–æ–∑ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
    predict = model.predict(features2)     
    
    print(f'== {model} ==')
    print()
    print(f'Predict F1 ({title}):', calc_F1(target2, predict))
```

```python
# –û–±—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –º–æ–¥–µ–ª–µ–π
def test_model2(model, features1, target1, features2, target2,
               flag_fit=True, flag_graph=True, title=''):
    if flag_fit:
        model = model.fit(features1, target1)
    predict = model.predict(features2)
    accuracy = accuracy_score(predict, target2)
    f1 = f1_score(target2, predict)
    recall = recall_score(target2, predict)
    precision = precision_score(target2, predict)
    predict_proba = model.predict_proba(features2)
    auc_roc = roc_auc_score(target2, predict_proba[:, 1])

    print(f'== {model.__class__} {title} ==')
    print('–ù–∞—Å—Ç—Ä–æ–π–∫–∏ –º–æ–¥–µ–ª–∏:', model)
    print('F1:', f1)
    print('Auc_roc:', auc_roc)
    print('Accuracy:', accuracy)
    print('Recall:', recall)
    print('Precision:', precision)
    print('–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫:\n', confusion_matrix(target2, predict))
    print()
    
    if flag_graph:
        fpr, tpr, thresholds = roc_curve(target2, predict_proba[:, 1])
        grafix_roc(fpr, tpr, title=f'{title}')
```
